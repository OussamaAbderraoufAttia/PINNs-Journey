# 09 - Literature Review & Paper Implementations

This notebook provides a comprehensive review of the PINN literature and key papers in scientific machine learning.

## Foundational Papers

### 1. Raissi, Perdikaris, & Karniadakis (2019)
**Physics-informed neural networks: A deep learning framework for solving forward and inverse problems involving nonlinear partial differential equations**

- **Key contribution**: First systematic framework for PINNs
- **Core idea**: Use automatic differentiation to enforce PDE residuals
- **Loss**: $\mathcal{L} = \mathcal{L}_{data} + \mathcal{L}_{physics}$
- **Applications**: Navier-Stokes, Schrödinger, Allen-Cahn equations

### 2. Lagaris, Likas, & Fotiadis (1998)
**Artificial neural networks for solving ordinary and partial differential equations**

- **Key contribution**: First use of neural networks for PDEs
- **Method**: Trial solution = particular solution + neural network
- **Limitations**: Limited to simple geometries

### 3. Psichogios & Ungar (1992)
**A hybrid neural network-first principles approach to process modeling**

- **Key contribution**: Hybrid physics-neural models
- **Idea**: Combine first-principles with data-driven components

## Architecture Improvements

### ResNet-based PINNs (Wang, S., et al., 2021)
**When and why PINNs fail to train: A neural tangent kernel perspective**

- **Problem**: PINNs can fail to train on stiff/multi-scale problems
- **Solution**: Residual connections improve gradient flow
- **Key insight**: Neural tangent kernel analysis explains training dynamics

### Fourier Feature Networks (Tancik et al., 2020)
**Fourier features let networks learn high-frequency functions in low-dimensional domains**

- **Problem**: MLPs with standard activations are biased toward low frequencies
- **Solution**: Random Fourier feature mappings as input
- **Application**: Capturing high-frequency wave solutions

### SIREN (Sitzmann et al., 2020)
**Implicit neural representations with periodic activation functions**

- **Innovation**: Periodic activation (sin) as primary nonlinearity
- **Advantage**: Better for modeling signals with sharp features
- **Use case**: Implicit neural representations, occupancy networks

## Training Strategies

### Adaptive Loss Weighting

1. **Gradient Normalization (GradNorm)** - Wang et al. (2021)
   - Balance gradients from different loss components
   - Adapt weights based on gradient magnitudes

2. **NTK-based weighting** - Wang et al. (2021)
   - Use Neural Tangent Kernel to determine importance
   - More principled than heuristic weighting

3. **Loss Balancing (Learning Rate Annealing)** - McClenny & Brailer (2020)
   - Adjust learning rates based on loss magnitudes
   - Simple but effective approach

### Adaptive Sampling

1. **RAR (Residual-based Adaptive Refinement)** - Lu et al. (2021)
   - Sample more where residuals are large
   - Focus computational effort on hard regions

2. **RAR-D (Density-based)** - Wang et al. (2021)
   - Combine residual magnitude with point density
   - Prevent clustering in easy regions

## Domain-Specific Applications

### Fluid Dynamics
- **DeepONet** (Lu et al., 2021): Operator learning for fluids
- **FNO** (Li et al., 2021): Fourier Neural Operator for Navier-Stokes
- **Causal PINNs** (Wang et al., 2022): Enforce causality in time

### Solid Mechanics
- **DeepSolid** (Samaniego et al., 2020): Solid mechanics with PINNs
- **Elasticity**: Stress-strain relations via automatic differentiation

### Electromagnetics
- **Maxwell's equations** via PINNs
- **Antenna design optimization**

### Quantum Mechanics
- **Schrödinger equation**: Ground state and time evolution
- **Quantum chemistry**: Electronic structure calculations

## Key Challenges & Open Problems

### 1. Multi-scale Problems
- **Issue**: Different scales in physics dominate differently
- **Solutions**: Multi-scale architectures, adaptive sampling

### 2. Stiff PDEs
- **Issue**: Rapidly changing solutions cause training instability
- **Solutions**: Causal training, curriculum learning

### 3. High-dimensional Problems
- **Issue**: Curse of dimensionality for sampling
- **Solutions**: Random projections, importance sampling

### 4. Boundary Condition Enforcement
- **Issue**: Hard vs soft BCs
- **Solutions**: Custom architectures, penalty methods

### 5. Uncertainty Quantification
- **Issue**: PINNs are deterministic
- **Solutions**: Bayesian PINNs, ensemble methods

## Next-Generation Approaches

### Neural Operators
- **DeepONet**: Operator networks with branch/trunk structure
- **FNO**: Fourier Neural Operator for learning mappings between function spaces
- **GNO**: Graph Neural Operators for unstructured domains

### Physics-Informed Transformers
- Attention mechanisms for long-range physics interactions
- Multi-head attention for different physical quantities

### Hybrid Approaches
- **Scientific ML**: Combining physics with data-driven methods
- **Theory-guided data science**: Physics as inductive bias
