# 01 - Autograd Fundamentals

```python
import torch, numpy as np
torch.manual_seed(42)
print(f'PyTorch: {torch.__version__}')
```

```python
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2
y.backward()
print(f'd/dx(x^2) at x=3: {x.grad.item()} (expected 6)')
```

```python
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
y = x ** 2
grad_y = torch.autograd.grad(y, x, grad_outputs=torch.ones_like(y))[0]
print(f'dy/dx = {grad_y} (expected [2,4,6])')
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x ** 3
dy = torch.autograd.grad(y, x, create_graph=True)[0]
d2y = torch.autograd.grad(dy, x, create_graph=True)[0]
d3y = torch.autograd.grad(d2y, x)[0]
print(f'y=x^3: dy/dx={dy.item()}, d2y/dx2={d2y.item()}, d3y/dx3={d3y.item()}')
```

```python
import torch.nn as nn
net = nn.Sequential(nn.Linear(2, 32), nn.Tanh(), nn.Linear(32, 32), nn.Tanh(), nn.Linear(32, 1))
xt = torch.tensor([[0.5, 0.3]], requires_grad=True)
u = net(xt)
g = torch.autograd.grad(u, xt, grad_outputs=torch.ones_like(u), create_graph=True)[0]
print(f'NN output: {u.item():.4f}, gradients: {g.detach()}')
```

```python
x = torch.tensor([[0.5, 0.5]], requires_grad=True)
u = torch.sin(np.pi * x[:, 0:1]) * torch.sin(np.pi * x[:, 1:2])
g = torch.autograd.grad(u, x, grad_outputs=torch.ones_like(u), create_graph=True)[0]
u_xx = torch.autograd.grad(g[:, 0:1], x, grad_outputs=torch.ones_like(g[:, 0:1]), create_graph=True)[0][:, 0:1]
u_yy = torch.autograd.grad(g[:, 1:2], x, grad_outputs=torch.ones_like(g[:, 1:2]), create_graph=True)[0][:, 1:2]
print(f'Laplacian of sin(pi*x)*sin(pi*y): {(u_xx+u_yy).item():.4f} (expected {-2*np.pi**2*np.sin(np.pi*0.5)**2:.4f})')
```
